```python
import sys
sys.path.append('..')
import numpy as np
from dezero import Variable

def rosenbrock(x0, x1):
    return 100 * (x1 - x0 ** 2) ** 2 + (1 - x0) ** 2
```

```python
x0 = Variable(np.array(0.0))
x1 = Variable(np.array(2.0))

y = rosenbrock(x0, x1)
y.backward()
```

```python
print(x0.grad, x1.grad)
```

```python
x0 = Variable(np.array(0.0))
x1 = Variable(np.array(2.0))
xs, ys = [], []
lr = 0.001
iters = 1000

import copy

for i in range(iters):
    print(x0, x1)
    
    y = rosenbrock(x0, x1)
    
    x0.cleargrad()
    x1.cleargrad()
    y.backward()
    
    x0.data -= lr * x0.grad
    x1.data -= lr * x1.grad
    xx = copy.deepcopy(x0.data)
    xy = copy.deepcopy(x1.data)
    
    xs.append(xx)
    ys.append(xy)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(xs, ys)
plt.show()
```
