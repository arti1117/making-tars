```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.arange(0, 6, .1)
y = np.sin(X)
```

```python
plt.plot(X, y)
plt.show()
```

```python
z = np.cos(X)
```

```python
plt.plot(X, y, label='sin')
plt.plot(X, z, linestyle='--', label='cos')
plt.xlabel("x")
plt.ylabel('y')
plt.title('sin & cos')
plt.legend()
plt.show()
```
