```python
import sys
sys.path.append('..')
import numpy as np
import matplotlib.pyplot as plt
from dezero import Variable
```

```python
def test_func(x):
    return x ** 4 - 2 * x ** 2
```

```python
x = np.arange(-2, 2, .00001)
y = test_func(x)
plt.plot(x, y)
plt.show
```

```python
def gx2(x):
    return 12 * x ** 2 - 4
```

```python
x = Variable(np.array(2.0))
iters = 10
```

```python
xs = []
ys = []
import copy
for i in range(iters):
    print(i, x)
    
    y = test_func(x)
    x.cleargrad()
    y.backward()
    xs.append(copy.deepcopy(x.data))
    ys.append(copy.deepcopy(y.data))
    
    x.data -= x.grad / gx2(x.data)
```

```python
x = np.arange(-2, 2, .00001)
y = test_func(x)
plt.plot(x, y)
plt.plot(xs, ys, marker=".")
plt.show()
```
