```python
import sys
sys.path.append('..')
import numpy as np
from dezero import Variable
import dezero.functions as F
```

```python
tx1 = Variable(np.array(1.0))
ty1 = F.square(tx1)
ty1.backward(create_graph=True)
```

```python
ty1.grad
tx1.grad
```

```python
tx2 = Variable(np.array(1.0))
ty2 = F.exp(tx2)
ty2.backward(create_graph=True)
```

```python
tx2.grad
```

```python
x = Variable(np.array(1.0))
y = F.tanh(x)
x.name = 'x'
y.name = 'y'
y.backward(create_graph=True)
```

```python
for i in range(0):
    gx = x.grad
    x.cleargrad()
    gx.backward(create_graph=True)
```

```python
gx = x.grad
gx.name = 'gx' + str(1)

from dezero.utils import plot_dot_graph

plot_dot_graph(gx, verbose=False, to_file='tanh.png')
```
